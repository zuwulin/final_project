Loading the Data

```python
#add dependencies
import pandas as pd
import numpy as np
```

```python
#file to load
gta_housing_csv = "GTA_houses_combined_precleaning.csv"

#read file and store into pandas dataframe
gta_housing_df = pd.read_csv(gta_housing_csv)

#check that everything loaded in
gta_housing_df.head(10)
```

Prettifying the Data

```python
#check column names
list(gta_housing_df)
```

```python
#delete first column as it is unneeded
gta_housing_df = gta_housing_df.iloc[: , 1:]

#check dataframe
gta_housing_df.head()
```

```python
#change column names for easier reading
gta_housing_df.columns = ["Listing_Image", "Listing_URL", "Full_Address", "City", "Zipcode", "Bedrooms", "Bathrooms", "Price", "Latitude", "Longitude", "Home_Type"]

#check dataframe again
gta_housing_df.head()
```

Starting Data Exploration and Data Cleaning

```python
#read dataframe into a different pandas dataframe
df = gta_housing_df

#check that everything loaded in
df.head(10)
```

```python
#get column types for dataframe of interest
df.info()
```

```python
#check for and drop any duplicates
df.drop_duplicates()

#no duplicates found
```

```python
#count values cities
df['City'].value_counts()
```

```python
#clean city columns if there is not enought data to test
indexCITY = df[df['City'] == 'Etobicoke'].index
df.drop(indexCITY , inplace=True)

indexCITY = df[df['City'] == 'Thornhill'].index
df.drop(indexCITY , inplace=True)

#check dataframe
df['City'].value_counts()
```

```python
#count values # of beds
df['Bedrooms'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexBED = df[df['Bedrooms'] == 12].index
df.drop(indexBED , inplace=True)

indexBED = df[df['Bedrooms'] == 9].index
df.drop(indexBED , inplace=True)

indexBED = df[df['Bedrooms'] == 0].index
df.drop(indexBED , inplace=True)

indexBED = df[df['Bedrooms'] == 6].index
df.drop(indexBED , inplace=True)

indexBED = df[df['Bedrooms'] == 7].index
df.drop(indexBED , inplace=True)

#check dataframe
df['Bedrooms'].value_counts()
```

```python
#count values # of baths
df['Bathrooms'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexBATH = df[df['Bathrooms'] == 13].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 10].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 11].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 0].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 9].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 8].index
df.drop(indexBATH , inplace=True)

indexBATH = df[df['Bathrooms'] == 7].index
df.drop(indexBATH , inplace=True)

#check dataframe
df['Bathrooms'].value_counts()
```

```python
#count values home type
df['Home_Type'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexHOME = df[df['Home_Type'] == 'MULTI_FAMILY'].index
df.drop(indexHOME , inplace=True)

indexHOME = df[df['Home_Type'] == 'LOT'].index
df.drop(indexHOME , inplace=True)

indexHOME = df[df['Home_Type'] == 'APARTMENT'].index
df.drop(indexHOME , inplace=True)

indexHOME = df[df['Home_Type'] == 'MANUFACTURED'].index
df.drop(indexHOME , inplace=True)

#check dataframe
df['Home_Type'].value_counts()
```

```python
#check if there are any abnormal house prices listed
df[df.Price<10000]
```

```python
#clean price column from abnormal prices
indexPRICE = df[df['Price'] < 1000].index
df.drop(indexPRICE , inplace=True)

#check dataframe
df['Price'].value_counts()
```

```python
#clean lat x long columns if their values are 0 as these will be useless for future dashboard
indexLAT = df[df['Latitude'] == 0].index
df.drop(indexLAT , inplace=True)

indexLONG = df[df['Longitude'] == 0].index
df.drop(indexLONG , inplace=True)

#check dataframe
df.info()
```

```python
#check the dataframe shape
df.shape

#number of entries after cleaning decreased from 2399 to 1851, by 548 entries
```
