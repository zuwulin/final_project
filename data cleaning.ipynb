```python
#add dependencies
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#file to load
gta_housing_csv = "resources/GTA_houses_combined_precleaning.csv"

#read file and store into pandas dataframe
gta_housing_df = pd.read_csv(gta_housing_csv)

#check that everything loaded in
gta_housing_df.head(10)
```

```python
gta_housing_df[gta_housing_df.price<1000]
```

```python
#check column names
list(gta_housing_df)
```

```python
#delete first column as it is unneeded
gta_housing_df = gta_housing_df.iloc[: , 1:]

#check dataframe
gta_housing_df.head()
```

```python
#change column names for easier reading
gta_housing_df.columns = ["listing image", "listing url", "full address", "city", "zipcode", "# of beds", "# of bathrooms", "price", "latitude", "longitude", "home type"]

#check dataframe again
gta_housing_df.head()
```

```python
#see overview of dataframe
gta_housing_df.info()
```

```python
#check for and drop any duplicates
gta_housing_df.drop_duplicates()
```

```python
#clean lat x long columns if their values are 0
indexLAT = gta_housing_df[ gta_housing_df['latitude'] == 0 ].index
gta_housing_df.drop(indexLAT , inplace=True)

indexLONG = gta_housing_df[ gta_housing_df['longitude'] == 0 ].index
gta_housing_df.drop(indexLONG , inplace=True)

#check dataframe
gta_housing_df.info()
```

```python
#count values cities
gta_housing_df['city'].value_counts()
```

```python
#clean city columns if there is not enought data to test
indexCITY = gta_housing_df[ gta_housing_df['city'] == 'Etobicoke' ].index
gta_housing_df.drop(indexCITY , inplace=True)

indexCITY = gta_housing_df[ gta_housing_df['city'] == 'Thornhill' ].index
gta_housing_df.drop(indexCITY , inplace=True)

#check dataframe
gta_housing_df['city'].value_counts()
```

```python
#count values zipcode
gta_housing_df['zipcode'].value_counts()
```

```python
#count values # of beds
gta_housing_df['# of beds'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexBED = gta_housing_df[ gta_housing_df['# of beds'] == 6 ].index
gta_housing_df.drop(indexBED , inplace=True)

indexBED = gta_housing_df[ gta_housing_df['# of beds'] == 0 ].index
gta_housing_df.drop(indexBED , inplace=True)

indexBED = gta_housing_df[ gta_housing_df['# of beds'] == 7 ].index
gta_housing_df.drop(indexBED , inplace=True)

indexBED = gta_housing_df[ gta_housing_df['# of beds'] == 12 ].index
gta_housing_df.drop(indexBED , inplace=True)

#check dataframe
gta_housing_df['# of beds'].value_counts()
```

```python
#count values # of baths
gta_housing_df['# of bathrooms'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 10 ].index
gta_housing_df.drop(indexBATH , inplace=True)

indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 14 ].index
gta_housing_df.drop(indexBATH , inplace=True)

indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 11 ].index
gta_housing_df.drop(indexBATH , inplace=True)

indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 13 ].index
gta_housing_df.drop(indexBATH , inplace=True)

indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 9 ].index
gta_housing_df.drop(indexBATH , inplace=True)

indexBATH = gta_housing_df[ gta_housing_df['# of bathrooms'] == 0 ].index
gta_housing_df.drop(indexBATH , inplace=True)

#check dataframe
gta_housing_df['# of bathrooms'].value_counts()
```

```python
#count values home type
gta_housing_df['home type'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexHOME = gta_housing_df[ gta_housing_df['home type'] == 'MULTI_FAMILY' ].index
gta_housing_df.drop(indexHOME , inplace=True)

indexHOME = gta_housing_df[ gta_housing_df['home type'] == 'LOT' ].index
gta_housing_df.drop(indexHOME , inplace=True)

indexHOME = gta_housing_df[ gta_housing_df['home type'] == 'APARTMENT' ].index
gta_housing_df.drop(indexHOME , inplace=True)

indexHOME = gta_housing_df[ gta_housing_df['home type'] == 'MANUFACTURED' ].index
gta_housing_df.drop(indexHOME , inplace=True)



#check dataframe
gta_housing_df['home type'].value_counts()
```

```python
#clean bedrooms column if there is not enought data to test
indexPRICE = gta_housing_df[ gta_housing_df['price'] < 1000 ].index
gta_housing_df.drop(indexPRICE , inplace=True)

indexPRICE = gta_housing_df[ gta_housing_df['price'] > 12998887 ].index
gta_housing_df.drop(indexPRICE , inplace=True)


#check dataframe
gta_housing_df['price'].value_counts()
```
